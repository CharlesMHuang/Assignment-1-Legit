```python
def create_X():
    Gamma_t = np.random.normal(0, 0.5, (240,6))
    Gamma_s = np.random.normal(0, 0.12247448713, (6,441))
    X = np.add(TC, Gamma_t).dot(np.add(SM, Gamma_s))
    X_avg = np.average(X, axis=0)
    X_sd = np.std(X, axis=0)

    COLS = 441
    # standardise
    for i in range(COLS):
        X[:,i] = X[:,i] - X_avg[i]
        X[:,i] = X[:,i] / X_sd[i]

    return X
```

```python
def create_A_LR(rho, X_interior):
    N = 240
    n_sources = 6
    x1 = 21
    x2 = 21

    step = 1/(np.linalg.norm(TC.dot(np.transpose(TC))) * 1.1)
    thr = rho * N * step
    Ao = np.zeros((n_sources, 1))
    A = np.zeros((n_sources, 1))
    A_LR = np.zeros((n_sources, x1 * x2))
    for k in range(x1 * x2):
        A = Ao + step * (np.transpose(TC).dot(X_interior[:,k:k+1] - TC.dot(Ao)))
        A = (1/(1+thr))*(np.sign(A)*np.maximum(np.zeros((6, 1)), np.absolute(A) - thr))
        for i in range(10):
            Ao = A
            A = Ao + step * (np.transpose(TC).dot(X_interior[:,k:k+1] - TC.dot(Ao)))
            A = (1/(1+thr))*(np.sign(A)*np.maximum(np.zeros((6, 1)), np.absolute(A) - thr))
        A_LR[:,k:k+1] = A

    return A_LR
```

```python
from statistics import mean

MSEs_per_rho = {}
for rho_big in range(5, 100, 5):
    rho = rho_big / 100
#     print(rho)
    MSEs_per_rho[rho] = []

# print(MSEs_per_rho)
    
for i in range(10):
    X_here = create_X()
    for rho_big in range(5, 100, 5):
        rho = rho_big / 100

        A_LR = create_A_LR(rho, X_here)
        D_LR = create_D(X_here, A_LR)
        MSEs_per_rho[rho].append(np.square(X - D_LR.dot(A_LR)).sum())

MSEs_per_rho = {k:mean(v) for k,v in MSEs_per_rho.items()}
print(MSEs_per_rho)    
```

```python
x = list(MSEs_per_rho.keys())
y = list(MSEs_per_rho.values())

plt.scatter(x,y)

plt.xlabel("rho")
plt.ylabel("Average MSE")

plt.show()
```
